### **Matplotlib**

A library that deals with graphs. Data Visualization tool

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

plt.scatter(x, y)
plt.plot(x, y)
```

```python
x1 = [1, 2, 3, 4, 5]
x2 = [1, 2, 3, 4, 5]

y1 = [1, 2, 3, 4, 5]
y2 = [5, 4, 3, 2, 1]

plt.scatter(x1, y1)
plt.plot(x1, y1)

plt.scatter(x2, y2)
plt.plot(x2, y2)
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

plt.scatter(x, y)
plt.plot(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
x2 = [1, 2, 3, 4, 5]

y = [1, 2, 3, 4, 5]
y2 = [5, 4, 3, 2, 1]

plt.scatter(x, y)
plt.plot(x, y, 'r')

plt.scatter(x2, y2)
plt.plot(x2, y2, 'g')

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
x2 = [1, 2, 3, 4, 5]

y = [1, 2, 3, 4, 5]
y2 = [5, 4, 3, 2, 1]

plt.scatter(x, y)
plt.plot(x, y, 'r--')

plt.scatter(x2, y2)
plt.plot(x2, y2, 'g')

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
x2 = [1, 2, 3, 4, 5]

y = [1, 2, 3, 4, 5]
y2 = [5, 4, 3, 2, 1]

plt.scatter(x, y)
plt.plot(x, y, 'r--')

plt.scatter(x2, y2)
plt.plot(x2, y2, 'g-.')

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
x2 = [1, 2, 3, 4, 5]

y = [1, 2, 3, 4, 5]
y2 = [5, 4, 3, 2, 1]

plt.scatter(x, y)
plt.plot(x, y, 'r*')

plt.scatter(x2, y2)
plt.plot(x2, y2, 'g+')

plt.show()
```

```python
import numpy as np
x = np.array([1, 2, 3, 4, 5, 6]) 
y = 2*(x**4) + 100
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = 2*(x**4) + 100
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = 5*(x**3) + 1000

plt.plot(x, y, color = "violet", marker = "*", linewidth = 1)
plt.show()
```

```python
x = np.arange(0, 10, 0.05)
y = 5*(x**3) + 1000
y2 = 5*(x**4) + 10

plt.plot(x, y, color = "red", label = "5*(x**3) + 1000")
plt.plot(x, y2, color = "black", label = "5*(x**4) + 10")
plt.legend()
```

```python
x = np.arange(0, 10, 0.05)
y = 5*(x**3) + 1000
y2 = 5*(x**4) + 10

plt.plot(x, y, color = "red", label = "5*(x**3) + 1000")
plt.plot(x, y2, color = "black", label = "5*(x**4) + 10")
plt.legend()
plt.xlabel("Natural Number", fontsize = 8)
plt.ylabel("Cube", fontsize = 8)
plt.axis([0, 20, 0, 10000])
plt.grid()
plt.text(15, 4000, "This prints the \ncube of a \nnumber", fontsize = 10)
plt.show()
```

```python
profit = [10, 20, 30, 40]
labels = ["Amazon", "Flipkart", "Snapdeal", "JioMart"]

plt.pie(profit)
plt.show()
```

```python
profit = [10, 20, 30, 40]
labels_list = ["Amazon", "Flipkart", "Snapdeal", "JioMart"]
colors_list = ["Red", "Yellow", "Blue", "Orange"]

plt.pie(profit, colors = colors_list, labels = labels_list, autopct = "%.2f%%")
plt.show()
```

```python
profit = [10, 20, 30, 40]
labels = ["Amazon", "Flipkart", "Snapdeal", "JioMart"]

colors = ["Red", "Yellow", "Blue", "Orange"]
explode = [0.05, 0.1, 0, 0]

plt.pie(profit, colors = colors, labels = labels, autopct = "%.2f%%", explode = explode)
plt.show()
```

```python
profit = [10, 20, 30, 40]
labels = ["Amazon", "Flipkart", "Snapdeal", "JioMart"]

colors = ["Red", "Yellow", "Blue", "Orange"]
explode = [0.05, 0.1, 0, 0]

plt.pie(profit, colors = colors, labels = labels, autopct = "%.2f%%", explode = explode, counterclock = False, startangle = 90)
plt.show()
```
