#**How to handle missing data**




```python
import pandas as pd
import numpy as np
```

```python
iris_df = pd.read_csv("/Iris.csv")
```

```python
data_frame = iris_df.copy()
```

```python
data_frame.describe()
```

```python
data_frame.head()
```

```python
# Adding missing values
data_frame.iloc[2:3, 2:4] = np.nan
data_frame.head()
```

```python
# Count of null values
data_frame.isnull().sum()
```

```python
# Removing null value rows
data_frame.dropna(inplace = True)
data_frame.head()
```

```python
data_frame.isnull().sum()
```

Resetting the Index

```python
# Resetting idex after removing null row
data_frame.reset_index(inplace = True)
data_frame.head()
```

Instead of writing all the above lines to reset the index, you can directly write this statement

```python
data_frame = iris_df.copy()
data_frame.iloc[2:3, 2:4] = np.nan
data_frame.dropna(inplace = True)
data_frame.head()
```

```python
#data_frame.reset_index(drop = True, inplace = True)
data_frame.head()
```

###**Find Mean, Mode and Median**

Replace the NaN values with either of the three values among mean, median and mode


```python
data_frame = iris_df.copy()
data_frame.iloc[2:3, 2:4] = np.nan
data_frame.head()
```

```python
# Fill null values with mean
data_frame.PetalLengthCm.fillna(data_frame.PetalLengthCm.mean(), inplace = True)
data_frame.head()
```
